```python
import kagglehub

# Download dataset
# Download latest version
path = kagglehub.dataset_download("blastchar/telco-customer-churn")

print("Path to dataset files:", path)

#
```

```python
import subprocess
from pathlib import Path
import shutil

# Current project directory
project_dir = Path.cwd()

print("Current project directory:", project_dir)

src_dir = Path(path)

# Run Linux cp command
subprocess.run(
    ["cp", "-r", f"{path}/.", str(project_dir)],
    check=True
)

print("Dataset copied to current project directory.")

# Delete original downloaded folder
shutil.rmtree(src_dir)

print("Original downloaded folder deleted.")
```
